```python
import numpy as np
```

```python
import pandas as pd

df=pd.read_csv()
```

```python
df.describe()
```

```python
skew = df.skew(numeric_only=True)
kurt = df.kurtosis(numeric_only=True)

eda_shape = pd.DataFrame({
    "skewness": skew,
    "kurtosis": kurt
}).sort_values("skewness", ascending=False)

print("\nSKEW & KURTOSIS:\n", eda_shape)
```

```python
q50 = df.quantile(0.50, numeric_only=True)
q99 = df.quantile(0.99, numeric_only=True)

tail_ratio = q99 / (q50 + 1e-9)

eda_tail = pd.DataFrame({
    "q99_to_median_ratio": tail_ratio
}).sort_values("q99_to_median_ratio", ascending=False)

print("\nHEAVY TAIL RATIO (q99 / median):\n", eda_tail)
```

```python
corr = df.corr(numeric_only=True)

# Flatten correlation matrix
corr_pairs = (
    corr.abs()
        .where(np.triu(np.ones(corr.shape), k=1).astype(bool))
        .stack()
        .sort_values(ascending=False)
)

print("\nTOP CORRELATED FEATURES (>0.9):\n", corr_pairs[corr_pairs > 0.9])
```
